**Q15: Using default dict, rank the tokenized sentences as per the frequency**

```python
# sentence that content most frequency word

from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
            
ranking
```

**Q18: Task1 : For whatever we just did above, make a function summarize so we input a text and n and get the top frequency sentences.**

```python
from heapq import nlargest
from nltk.probability import FreqDist
from collections import defaultdict

def summarize(text, n):
    sents = sent_tokenize (text)
    
    words = word_tokenize (text.lower())
    _stopwords = set(stopwords.words('english') + list(punctuation))
    words = [w for w in words if w not in _stopwords]
    
    freq_dict = FreqDist(words)
    topfreq = nlargest(n, freq_dict, key = freq_dict.get)

    ranking = defaultdict(int)
    
    for index, se in enumerate(sents): 
        for w in word_tokenize(se.lower()):
            if w in topfreq:
                ranking[index] += freq_dict[w]
    top_five_index = sorted (nlargest(5,ranking,key = ranking.get))
    summary = [sents[index] for index in top_five_index]      
    return summary
```

```python
result = summarize (text,20)
```
